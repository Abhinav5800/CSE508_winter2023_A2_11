```python
c=[]
for i in range(1,137):
    a="Feature_"+str(i)
    c.append(a)
```

```python
def features(row):
    for column in c:
        row[column] = row[column].split(':')[1]
        row[column] = float(row[column])
    return row
```

```python
df2 = df2.apply(features,axis='columns')
df2
```

Third objective

```python
df2_sorted_f_75=df2_sorted.sort_values(by="Feature_75",ascending=False)
df2_sorted_f_75
```

```python
doc = df2_sorted_f_75.query('Relevance_score > 0')
docs = doc.shape[0]
doc
```

```python
print(df2_sorted_f_75.iterrows())
```

```python
precision_list = []
recall_list = []
count = 1
relevant_docs = 0

for index,row in df2_sorted_f_75.iterrows():
    if df2_sorted_f_75.loc[index]['Relevance_score'] > 0:
        relevant_docs += 1
    
    precision = relevant_docs / count
    recall = relevant_docs / docs

    precision_list.append(precision)
    recall_list.append(recall)
    count+=1

df2_sorted_f_75['precision'] = precision_list
df2_sorted_f_75['recall'] = recall_list
```

```python
precision =  df2_sorted_f_75['precision']
recall = df2_sorted_f_75['recall']
```

```python
import matplotlib.pyplot as plt
plt.plot(recall,precision)
plt.title('Precision-Recall curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.show()
```
